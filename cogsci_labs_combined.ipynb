8. Let's deal with the missing values. 

a) For the features EthnicGroup, ParentMaritalStatus and WklyStudyHours replace NaN values with the most common value.

b) For the feature ParentEduc replace NaN value with a string *no formal education*.

c) For the feature TestPrep replace NaN value with a string *none*.

d) For the features PracticeSport and IsFirstChild remove all the examples with NaN values. 

e) For the feature NrSiblings replace the NaN value with the median value.

f) Finally, for the feature TransportMeans replace the NaN value with a string *walking*.




```python
#a
mostCommonEthnicGroup = df["EthnicGroup"].mode()[0]
print(mostCommonEthnicGroup)
df["EthnicGroup"].fillna(mostCommonEthnicGroup, inplace=True)

mostCommonParentMaritalStatus = df["ParentMaritalStatus"].mode()[0]
print(mostCommonParentMaritalStatus)
df["ParentMaritalStatus"].fillna(mostCommonParentMaritalStatus, inplace=True)

mostCommonWeeklyStudyHours = df["WklyStudyHours"].mode()[0]
print(mostCommonWeeklyStudyHours)
df["WklyStudyHours"].fillna(mostCommonWeeklyStudyHours, inplace=True)

#b
df["ParentEduc"].fillna("no formal education", inplace=True)

#c
df["TestPrep"].fillna("none", inplace=True)

#d
df.dropna(subset=["PracticeSport"], inplace=True)
df.dropna(subset=["IsFirstChild"], inplace=True)

df.reset_index(drop=True, inplace=True)

#e
med = df["NrSiblings"].median()
df["NrSiblings"].fillna(med, inplace=True)

#f
df["TransportMeans"].fillna("walking", inplace=True)

print(df.isna().sum())
```

3. a) Use a statistical test to test the hypothesis:<br>

```python
import scipy.stats as stats
print(stats.ttest_ind(df.loc[df["Gender_male"] == 1]["MathScore"], df.loc[df["Gender_female"] == 1]["MathScore"]))
```

4. a) Use a statistical test to test the hypothesis:<br>
   H0 Gender did not affect the writing scores.<br>
   H1 Gender affected the writing score.<br>

```python
import scipy.stats as stats

male_scores = df.loc[df["Gender_male"] == 1]["WritingScore"]
female_scores = df.loc[df["Gender_female"] == 1]["WritingScore"]

f_statistic, p_value = stats.f_oneway(male_scores, female_scores)

print("ANOVA F-statistic:", f_statistic)
print("ANOVA p-value:", p_value)

```

**Gender affected the writing score.**

2. Explain the differences between in-house data labeling, crowdsourcing and outsourcing. 

We choose which method to use depending on the problem we are faced with, but also based on how much time and people we have at our disposal.

In-house labeling will get us the best-quality outcome. Correct labeling can be crucial in certain situations, e.g. when it comes to insurance or healthcare. This is why in-house labeling often requires us to consult experts in the field in order to correctly label the data. However, this process can be extremely slow.

On the other hand, crowdsorcing means that we seek help from a large number of freelancers. Some crowdsourcing platforms have up to tens of thousands of registered data annotators at their disposal at any given moment. These annotators will most often be hired to work on trivial data that does not require additional expertise in a particular field, e.g. images of animals, plants, environment. 

Outsourcing is somewhere in-between. These annotators will not be random freelancers, but people who were trained by an individual or an organization. This then means that we can evaluate the annotators on a specific topic before we hand them over the work. With outsourcing, we can ensure good quality of data marking without spending too much money.

3. Explain the main differences between annotation for image classification, object detection and image segmentation.

Image classification consists of adding tags to images. There is a certain number of classes that a model can classify, and this number corresponds to the number of unique tags found in the database. This classification can either be binary or multi-class (containing several labels). It is also possible to classify pictures with multiple tags.

The task of object detection, as the name suggests, is to detect different objects and determine their locations. Each detected object will be encapsulated with a bounding box that marks its reach. A label is usually attached to each of these boxes. The coordinates and the label of the box will be stored in a JSON file in the dictionary format.

Image segmentation is the process of seperating objects in the image from their background, as well as from other objects found in the image. The process usually results in images of the same size that will contain 1 where the object is present and 0 where the object is not present. If more than one object is segmented in a particular image, each of the objects will be marked on a seperate channel. The sum of these channels will then constitute the reference image.

The main task of this notebook is to make predictions using supported learning with the Scikit Learn library. The goal is to predict the grade in Math considering other features of the student.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
```

1. Load the cleaned dataset from the first exercise. If you haven't saved the dataset, rerun the exercise and save the final dataset.

```python
df = pd.read_csv("students_score_cleaned.csv")
df.drop(columns=["Unnamed: 0"], inplace=True)
df
```

2. Transform the MathScore feature into grades (1-5) using the following scoring system:

|   MathScore |  Grade  |
| ----------- | ------- |
|  88 - 100   |    5    |
|  75 - 87    |    4    |
|  63 - 74    |    3    |
|  50 - 62    |    2    |
|   0 - 49    |    1    |

```python
def grade(score):
    if score >= 88:
        return 5
    elif score >= 75:
        return 4
    elif score >= 63:
        return 3
    elif score >= 50:
        return 2
    else:
        return 1

df["MathScore"] = df["MathScore"].apply(grade)
df
```

3. Divide the dataset into features (X) and predictions (y). For features we will use all the columns except MathScore, ReadingScore and WritingScore. For predictions we will use the MathScore column. Split the dataset into training and testing sets. The split should be done in aÂ  70-30% ratio.

```python
X = df.drop(columns=["MathScore", "ReadingScore", "WritingScore"])
y = df[["MathScore"]]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
y_train
```

4. Create a Random Forest model, with max_depth=5 and n_estimators=20. Train the model using training set, and then test it on testing set. Display the confusion matrix. Show precision, recall and F1 score for all grades. 

```python
model = RandomForestClassifier(n_estimators=20, max_depth=5)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, y_pred))
disp.plot()
plt.show()

print(classification_report(y_test, y_pred))
```

5. The accuracy of our model is not the best. The reason could be having too many possible classes (grades).
